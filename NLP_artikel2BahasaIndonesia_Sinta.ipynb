## Preproses Text dengan NLTK dan Sastrawi

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
f = open('artikel_2.txt', 'r', encoding='utf-8')
isi_artikel = f.read()
print(isi_artikel)
```

```python
wt = word_tokenize(isi_artikel)
print(wt)
```

```python
sw = set(stopwords.words('indonesian'))
wtns = [w for w in wt if not w in sw]
print(wtns)
```

```python
freq_kt_1 = nltk.FreqDist(wt)
freq_kt_2 = nltk.FreqDist(wtns)
print(freq_kt_1.most_common(25))
print()
print(freq_kt_2.most_common(25))
```

```python
import matplotlib.pyplot as plt
```

```python
freq_kt_1.plot(25)
freq_kt_2.plot(25)
plt.show()
```
