# Preproses text dengan NLTK Sastrawi

```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
df = open('Artikel_Bahasa_Indonesia.txt', 'r')
isi_artikel = df.read()
print(isi_artikel)
```

```python
word_tokens = word_tokenize(isi_artikel)
print(word_tokens)
```

```python
stop_words = set(stopwords.words('indonesian'))
word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
print(word_tokens_no_stopwords)
```

```python
freq_kata_1 = nltk.FreqDist(word_tokens)
freq_kata_2 = nltk.FreqDist(word_tokens_no_stopwords)
print(freq_kata_1.most_common(25))
print()
print(freq_kata_2.most_common(25))
```

```python
import matplotlib.pyplot as plt
```

```python
freq_kata_1.plot(25)
freq_kata_2.plot(25)
plt.show()
```
