```python
vader_sentiment = SentimentIntensityAnalyzer()
data['scores_VADER'] = data['Text'].apply(lambda s: vader_sentiment.polarity_scores(s)['compound'])
```

```python
data['pred_VADER'] =data['scores_VADER'].apply(lambda x: 1 if x>=0 else 0)
data.head()
```
