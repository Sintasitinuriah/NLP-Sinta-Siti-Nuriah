# Menampilkan Naskah pidato asli

```python
pidato20 = open('./drive/MyDrive/UTS-NLP-Sinta-Siti-Nuriah-10219008/pidato_presiden_21.txt')
isi_pidato = pidato20.read()
print(isi_pidato)
```

# Membersihkan data dari tanda baca dan mengubah teks ke dalam bentuk lower case dan menampilkan penggalan kata sebelum dihilangkan kata konjungsinya

```python
import string
cleaning_text = isi_pidato.translate(str.maketrans('','', string.punctuation)).lower()
word_tokens = word_tokenize(cleaning_text)
print(word_tokens)
```

# Import string dan Menampilkan penggalan kata yang telah dihilangkan kata konjungsinya

```python
import string
stop_words = set(stopwords.words('indonesian'))
word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
print(word_tokens_no_stopwords)
```

```python
# kemunculan setiap token yang belum difiltering
freq_kata1 = nltk.FreqDist(word_tokens)
# kemunculan setiap token yang telah difiltering
freq_kata2 = nltk.FreqDist(word_tokens_no_stopwords)
# Menampilkan token yang paling sering muncul sebelum difiltering
print(freq_kata1.most_common(10))
print()
# Menampilkan token yang paling sering muncul setelah difiltering
print(freq_kata2.most_common(10))
```

# Plot 10 kata paling sering muncul sebelum data dibersihkan

```python
import matplotlib.pyplot as plt
freq_kata1.plot(10)
plt.show()
```

# Plot 10 kata paling sering muncul setelah data dibersihkan

```python
freq_kata2.plot(10)
plt.show()
```
